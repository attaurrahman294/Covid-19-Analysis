```python
# Remove the records where state is india
vaccine = data2_var[data2_var.State!='India']
vaccine.head()
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated': 'Total'}, inplace=True)
vaccine.head()
```

```python
#most number of vaccinated state
most_vac = vaccine.groupby('State')['Total'].sum().to_frame('total')
most_vac = most_vac.sort_values('total', ascending=False)[:5]
most_vac
```
